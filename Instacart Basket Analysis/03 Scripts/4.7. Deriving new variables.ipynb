```python
# Import libraries
import pandas as pd
import numpy as np
import os
```

```python
# Import data set 
df_ords_prods_merged = pd.read_pickle(r'/Users/OldBobJulia/Desktop/CF/Course/4. Python/Instacart Basket Analysis/02 Data/Prepared data/orders_products_merged.pkl')
```

```python
# Create subset of 1000000
df = df_ords_prods_merged[:1000000]
```

```python
df.shape
```

```python
# Create price ranges for product with user-defined function (uses a lot of RAM!)
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Check head
df.head()
```

```python
# Count values
df['price_range'].value_counts()
```

```python
# Create product range variable on full data set using .loc
df_ords_prods_merged.loc[df_ords_prods_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['prices'] <= 15) & (df['prices'] > 15), 'price_range_loc'] = 'Mid-range product' 
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df_ords_prods_merged.head()
```

```python
# Count values
df_ords_prods_merged['price_range_loc'].value_counts(dropna=False)
```

```python
# Count orders during each day of the week
df_ords_prods_merged['orders_day_of_week'].value_counts(dropna=False)
```

```python
# Create loop to classify day of the week by business 
result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# Check result
result
```

```python
# Create new column with result
df_ords_prods_merged['busiest_day'] = result 
```

```python
df_ords_prods_merged.head()
```

```python
df_ords_prods_merged.shape
```

```python
# Count values new column
df_ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

```python
# Drop previous busiest day column to correct it as per business question
df_ords_prods_merged.drop(columns = ['busiest_day'])
```

```python
# Create loop for 2 busiest days and 2 slowest days
result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0 or value == 1:
    result.append("Busiest days")
  elif value == 4 or value == 3:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# Create new column with result_2
df_ords_prods_merged['busiest_day'] = result 
```

```python
df_ords_prods_merged.head()
```

```python
# Count busiest periods of the day 
df_ords_prods_merged['order_hour_of_day'].value_counts(dropna=False)
```

```python
# Create for loop for busiest and least busy hours of the day 
result = []

for value in df_ords_prods_merged["order_hour_of_day"]:
  if value == 10 or value == 11 or value == 14 or value == 15 or value == 13:
    result.append("Most orders")
  elif value == 3 or value == 4 or value == 2 or value == 5 or value == 1:
    result.append("Fewest orders")
  else:
    result.append("Average orders")
```

```python
# Create new column
df_ords_prods_merged['busiest_period_of_day'] = result
```

```python
# Check counts new column
df_ords_prods_merged['busiest_period_of_day'].value_counts(dropna = False)
```
